```python
import requests
from bs4 import BeautifulSoup
import pickle
import numpy as np
```

```python
def GetPageContent(url):
    page = requests.get(url, headers={"Accept-Language": "en-US"})
    return BeautifulSoup(page.text, "html.parser")
def CrawlData(url):
    print(f"Crawling URL: {url}")
    soup = GetPageContent(url)
    paragraphs = soup.find('div', id = 'maincontent')
    if paragraphs is None:
        print(f"Failed to get paragraphs: {url}")
        return
    paragraphs = paragraphs.find_all('p')
    return ' '.join([p.get_text(strip = True) for p in paragraphs])
```

```python
with open('urllink.txt', 'r') as f:
    urllink = [line.strip() for line in f.readlines()]

# Tạo danh sách để lưu dữ liệu
data = []
for url in urllink:
    content = CrawlData(url)
    if content:
        data.append(content)
```

```python
data = np.array(data)
pickle.dump(data, open('crawling.sav', 'wb'))
```
